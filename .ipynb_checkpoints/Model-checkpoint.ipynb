### 3.Splitting the data into train set and test set

```python
from sklearn.model_selection import train_test_split
X = np.array(Data.drop(columns = ['charges']))
y = np.array(Data['charges']).reshape(-1,1)
print(X.shape, '\n',
y.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.15,random_state=42)
```

```python
X_train[0:6,:]
```

### 4. Creating the model

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_train)
y_pred.shape
```

```python
print('Coefficients are: \n', model.coef_)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

#Mean squared error
print('Mean Squared Error: \n',mean_squared_error(y_train,y_pred))
print('Coefficient of determination: \n', r2_score(y_train,y_pred))
```

```python
# Testing our model

test_pred = model.predict(X_test)
```

```python
print('Mean Squared Error: \n',mean_squared_error(y_test,test_pred))
print('Coefficient of determination: \n', r2_score(y_test,test_pred))
```

```python
print('Score of the model on training set: ',model.score(X_test, y_test))
```

Score of Linear Model is nothing but R2 score

### 5. Making a predictive Model

```python
input_data = list(map(float,input().split()))

#Changing the input_data to a numpy array
data_array = np.array(input_data, dtype = float)
```
